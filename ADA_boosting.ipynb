```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
```

```python
X,y  = make_classification(n_samples=1000,n_features=20,n_classes = 2,random_state=2)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
classifier = AdaBoostClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)

#Evaluation
print("current model performance : \n")
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators' : [50,100,150,200,250,],
    'learning_rate' : [0.001,0.01,0.1,1.0],
    'algorithm' : ['SAMME','SAMME.R']
}

ada = AdaBoostClassifier()
grid_search = GridSearchCV(estimator=ada,param_grid=param_grid,cv=5,n_jobs=-1,verbose=1)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)

#Evaluation
print("current model performance : \n")
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#adaboost regressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.ensemble import AdaBoostRegressor
```

```python
X,y = make_regression(n_samples=1000,n_features=2,noise=10,random_state=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
regressor = AdaBoostRegressor()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)

#Evaluation
print("current model performance : \n")
print(f"R2 SQUARE SCORE : {r2_score(y_test,y_pred)}")
print(f"mse score : {mean_squared_error(y_test,y_pred)}")
print(f" mean absolute error score : {mean_absolute_error(y_test,y_pred)}")
```
